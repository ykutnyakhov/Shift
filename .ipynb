```python
# факториал числа 2024
import math

def factorial(n):
    return math.factorial(n)

print("факториал числа 2024:", factorial(2024))
```

```python
# среднее 1000 случайных чисел [0, 100]

import random

numbers = []

for _ in range(1000):
    numbers.append(random.randint(0, 100))

print("среднее 1000 случайных чисел [0, 100]:", sum(numbers) / len(numbers))
```
